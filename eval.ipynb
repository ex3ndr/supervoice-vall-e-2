```python
import torch
from IPython.display import Audio, display
```

# Loading model from Torch Hub

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = torch.hub.load(repo_or_dir='ex3ndr/supervoice-vall-e-2', model='supervoice')
model = model.to(device)
```

```python
predicted = model.synthesize("voice_2", "What time is it, Steve?", top_p = 0.2).cpu()
display(Audio(data=predicted.cpu(), rate=24000))
```
